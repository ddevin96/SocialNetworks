```julia
using Pkg
Pkg.add("CSV")
Pkg.add("DataFrames")
Pkg.add("Plots")
using CSV
using DataFrames
using Plots
```

```julia
# DataFrame(CSV.File("../../results/mts/ca-GrQc.edges.txt"))

# read all content in a folder
df = DataFrame()
for dir in readdir("../../results")
    println(dir)
    for file in readdir("../../results/$dir")
        # if file ends with .txt
        if !occursin(".txt", file)
            continue
        end

        df_temp= DataFrame(CSV.File("../../results/$dir/$file"))
        # add df_temp to df
        df = vcat(df, df_temp)
    end
end
a = ["Algorithm", "Graph", "l", "th", "avg"]
rename!(df, names(df) .=> a)
println(df)

graphs = ["karate", "ca-GrQc", "ca-HepTh", "CollegeMsg", "email-EU-core", "facebook", "p2p"]
```
